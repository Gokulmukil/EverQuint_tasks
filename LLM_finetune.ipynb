```python
from datasets import load_dataset
import pandas as pd

dataset = load_dataset(
    "wikimedia/wikipedia",
    "20231101.en",
    split="train"
)

corpus = dataset.select(range(10000))
df = pd.DataFrame(corpus)

print(f"Loaded {len(df)} documents")
print(df["text"][0][:200])
```

```python
import re
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from bs4 import BeautifulSoup  # For any HTML-like wiki tags

# Download NLTK data (for completeness)]
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('omw-1.4', quiet=True)  # Better lemmatization

# Setup
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_wiki_text(text):
    if pd.isna(text) or not isinstance(text, str):
        return ""

    # Step 1: Strip common Wikipedia markup
    text = re.sub(r'\[\[([^\]]*)\]\]', r'\1', text)  # Simplify links
    text = re.sub(r'{{[^}}]*}}', '', text)  # Templates
    text = re.sub(r'\{\|[^}]*\}', '', text)  # Infoboxes/tables
    text = re.sub(r'<ref[^>]*>.*?</ref>', '', text, flags=re.DOTALL)  # References
    text = re.sub(r'<!--.*?-->', '', text, flags=re.DOTALL)  # Comments
    text = re.sub(r'\[\d+\]', '', text)  # Citations

    # Step 2: HTML/Wiki tags
    soup = BeautifulSoup(text, 'html.parser')
    text = soup.get_text()

    # Step 3: Normalize
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = text.lower()

    # Step 4: Tokenize, filter, lemmatize
    words = nltk.word_tokenize(text)
    words = [lemmatizer.lemmatize(w) for w in words if w not in stop_words and len(w) > 2]

    return ' '.join(words)

# Apply cleaning (your df from load_dataset)
print("Cleaning started... (5-10 mins for 10k docs)")
df['cleaned_text'] = df['text'].apply(clean_wiki_text)
df['title_clean'] = df['title'].apply(lambda x: clean_wiki_text(str(x)) if pd.notna(x) else "")

# Filter short full docs first
df = df[df['cleaned_text'].str.len() > 100].reset_index(drop=True)
print(f"After doc filter: {len(df)} docs")

# Add doc ID before chunking (will duplicate on explode)
df['original_doc_id'] = df.index.tolist()

# Chunk long texts
def chunk_text(text, max_words=500):
    if not text:
        return []
    words = text.split()
    return [' '.join(words[i:i+max_words]) for i in range(0, len(words), max_words)]

df['chunks'] = df['cleaned_text'].apply(chunk_text)

# Explode to one row per chunk (duplicates metadata)
df_exploded = df.explode('chunks').reset_index(drop=True)

# Filter short chunks *after* explode
df_exploded = df_exploded[df_exploded['chunks'].str.len() > 100].reset_index(drop=True)

# No manual lists needed—'title' and 'original_doc_id' are already duplicated!

print(f"Original docs: {len(df)}, Total chunks after filtering: {len(df_exploded)}")
print(f"Avg chunks per doc: {len(df_exploded) / len(df):.1f}")
print(f"Sample chunk: {df_exploded['chunks'][0][:200]}...")
print(f"Sample title: {df_exploded['title'][0]}")
print(f"Sample doc ID: {df_exploded['original_doc_id'][0]}")

# Save
df_exploded.to_json('corpus_cleaned.json', orient='records', lines=True)
print("Saved to corpus_cleaned.json")
```

```python
# Install if needed (run once)
!pip install sentence-transformers faiss-cpu rank-bm25 -q

import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
import faiss
from rank_bm25 import BM25Okapi
import pickle  # For saving index/models

# Load your cleaned corpus
df_exploded = pd.read_json('corpus_cleaned.json', lines=True)
texts = df_exploded['chunks'].tolist()
print(f"Loaded {len(texts)} chunks for indexing")

# Step 1: Generate embeddings (lightweight model: fast, good for English wiki)
model = SentenceTransformer('all-MiniLM-L6-v2')  # 384-dim, ~22MB
print("Encoding texts... (2-5 mins)")
embeddings = model.encode(texts, batch_size=64, show_progress_bar=True, convert_to_numpy=True)

# Normalize for cosine similarity
faiss.normalize_L2(embeddings)

# Step 2: Build FAISS index
dimension = embeddings.shape[1]
index = faiss.IndexFlatIP(dimension)  # Inner product = cosine sim for normalized vecs
index.add(embeddings.astype('float32'))
faiss.write_index(index, 'wiki_embeddings.index')
print(f"Indexed {len(texts)} vectors (dim: {dimension})")

# Step 3: BM25 index (for hybrid rerank)
tokenized_corpus = [text.split() for text in texts]
bm25 = BM25Okapi(tokenized_corpus)
with open('bm25_index.pkl', 'wb') as f:
    pickle.dump(bm25, f)
print("BM25 index saved")

# Quick test: Embed a sample query
sample_query = "anarchism philosophy"
query_emb = model.encode([sample_query])
faiss.normalize_L2(query_emb)
scores, idx = index.search(query_emb.astype('float32'), k=3)
print(f"Top 3 semantic matches for '{sample_query}':")
for i, score in zip(idx[0], scores[0]):
    print(f"  Score: {score:.3f} | Chunk: {texts[i][:100]}...")
```

```python
# Load indices (if restarted)
model = SentenceTransformer('all-MiniLM-L6-v2')
index = faiss.read_index('wiki_embeddings.index')
with open('bm25_index.pkl', 'rb') as f:
    bm25 = pickle.load(f)
texts = df_exploded['chunks'].tolist()  # Reload if needed
titles = df_exploded['title'].tolist()
doc_ids = df_exploded['original_doc_id'].tolist()

def search_documents(query, n=5, alpha=0.7):
    """
    Hybrid search: Embeddings (semantic) + BM25 (keywords), weighted rerank.
    alpha: Weight for embeddings (0.7 = more semantic).
    Returns: List of (chunk_text, title, doc_id, score)
    """
    # Semantic: FAISS cosine sim
    query_emb = model.encode([query])
    faiss.normalize_L2(query_emb)
    emb_scores, emb_indices = index.search(query_emb.astype('float32'), k=n*2)  # Over-retrieve for rerank

    # Keyword: BM25
    tokenized_query = query.lower().split()
    bm25_scores = bm25.get_scores(tokenized_query)

    # Rerank: Weighted avg of scores (normalize BM25 roughly)
    bm25_norm = np.array(bm25_scores)[emb_indices[0]] / (np.array(bm25_scores).max() + 1e-8)
    combined_scores = alpha * emb_scores[0] + (1 - alpha) * bm25_norm

    # Top N by combined
    top_idx = np.argsort(combined_scores)[::-1][:n]
    results = []
    for i in top_idx:
        doc_idx = emb_indices[0][i]
        results.append({
            'text': texts[doc_idx],
            'title': titles[doc_idx],
            'doc_id': doc_ids[doc_idx],
            'score': combined_scores[i]
        })
    return results

# Test with 3 queries
test_queries = [
    "anarchism philosophy",  # Should hit your sample
    "climate change effects",
    "artificial intelligence ethics"
]

for q in test_queries:
    print(f"\n--- Query: '{q}' ---")
    results = search_documents(q, n=3)
    for r in results:
        print(f"Score: {r['score']:.3f} | Title: {r['title'][:50]}... | Doc ID: {r['doc_id']}")
        print(f"Excerpt: {r['text'][:150]}...")
```

```python
# Cell 2: Load & Prep Dataset
import pandas as pd
from datasets import Dataset
import numpy as np

# Load your cleaned corpus
df = pd.read_json('corpus_cleaned.json', lines=True)
print(f"Loaded {len(df)} chunks")

# Generate pseudo-gold summaries: Extractive (first 20% words) + ellipsis for abstractive feel
def make_gold_summary(text, ratio=0.2):
    words = text.split()
    summary_len = max(30, int(len(words) * ratio))  # Min 30 words
    return ' '.join(words[:summary_len]) + ' ... (continued in full article)'

df['summary'] = df['chunks'].apply(make_gold_summary)

# Sample 5k for training (diverse: random sample)
train_df = df.sample(n=5000, random_state=42).reset_index(drop=True)
train_df = train_df[['chunks', 'summary']].rename(columns={'chunks': 'text', 'summary': 'summary_text'})

# Split: 90/10 train/val
split = int(0.9 * len(train_df))
train_ds = Dataset.from_pandas(train_df[:split])
val_ds = Dataset.from_pandas(train_df[split:])

dataset = {'train': train_ds, 'test': val_ds}  # 'test' for val
dataset['train'].save_to_disk('wiki_summary_train')
dataset['test'].save_to_disk('wiki_summary_val')
print(f"Train: {len(dataset['train'])}, Val: {len(dataset['test'])}")
print("Sample pair:\nInput:", dataset['train'][0]['text'][:100] + "...\nSummary:", dataset['train'][0]['summary_text'])
```

```python
 # Add this at the top with other imports
from transformers import (
    AutoTokenizer, AutoModelForSeq2SeqLM, Seq2SeqTrainer, Seq2SeqTrainingArguments,
    DataCollatorForSeq2Seq
)
from peft import LoraConfig, get_peft_model  # Removed TaskType import
from datasets import load_from_disk
import torch
import numpy as np  # Added for compute_metrics
from rouge_score import rouge_scorer

# Load data
train_dataset = load_from_disk('wiki_summary_train')
val_dataset = load_from_disk('wiki_summary_val')

# Model & Tokenizer
model_name = "facebook/bart-large-cnn"  # Proven for summarization
tokenizer = AutoTokenizer.from_pretrained(model_name)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
model = AutoModelForSeq2SeqLM.from_pretrained(
    model_name,
    # load_in_4bit=True,  # Quantization: Cuts VRAM ~75%
    device_map="auto",
    dtype=torch.float16 # Changed from torch_dtype to dtype
)

# LoRA Config
lora_config = LoraConfig(
    # task=TaskType.SEQ_2_SEQ_LM,  # Add this back
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    target_modules=["q_proj", "v_proj", "k_proj", "out_proj"]
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()  # Expect ~1M trainable params

# Preprocess
def preprocess_function(examples):
    inputs = tokenizer(examples['text'], max_length=1024, truncation=True, padding=False)
    labels = tokenizer(examples['summary_text'], max_length=128, truncation=True, padding=False)
    inputs['labels'] = labels['input_ids']
    return inputs

train_dataset = train_dataset.map(preprocess_function, batched=True, remove_columns=train_dataset.column_names)
val_dataset = val_dataset.map(preprocess_function, batched=True, remove_columns=val_dataset.column_names)

# Data collator
data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model, padding=True)

# Custom metrics (ROUGE)
def compute_metrics(eval_preds):
    preds, labels = eval_preds
    if isinstance(preds, tuple):
        preds = preds[0]
    decoded_preds = tokenizer.batch_decode(preds, skip_special_tokens=True)
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True)
    rouge1 = [scorer.score(t, p)['rouge1'].fmeasure for t, p in zip(decoded_labels, decoded_preds)]
    rougeL = [scorer.score(t, p)['rougeL'].fmeasure for t, p in zip(decoded_labels, decoded_preds)]
    return {'rouge1': sum(rouge1)/len(rouge1), 'rougeL': sum(rougeL)/len(rougeL)}

# Training args
training_args = Seq2SeqTrainingArguments(
    output_dir='./bart_wiki_finetuned',
    eval_strategy="epoch",
    save_strategy="epoch",  # Add this line
    learning_rate=3e-4,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=4,
    num_train_epochs=2,
    weight_decay=0.01,
    save_total_limit=2,
    predict_with_generate=True,
    generation_max_length=128,
    fp16=True,
    report_to=None,
    load_best_model_at_end=True,
    metric_for_best_model="rougeL"
)



# Trainer
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

# Train!
print("Starting fine-tuning...")
trainer.train()
trainer.save_model('./bart_wiki_finetuned')
tokenizer.save_pretrained('./bart_wiki_finetuned')
print("Fine-tuning complete! Model saved.")
print("Final Val Metrics:", trainer.evaluate())
```
